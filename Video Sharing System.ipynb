```python
weights={'view':0.2,'like':0.4,'inspire':0.4,'rating':0.1,'upvote_count':0.4,'view_count':0.2,'rating_count':0.01}
combined_data['interaction_score']=(
    combined_data['view']*weights['view']+combined_data['like']*weights['like']+combined_data['inspire']*weights['inspire']+combined_data['rating']*weights['rating']
    +combined_data['upvote_count']*weights['upvote_count']+combined_data['view_count']*weights['view_count']+combined_data['rating_count']*weights['rating_count']

    )
```

```python
user_post_matrix=combined_data.pivot_table(
    index="user_id_y",columns="post_id",values="interaction_score",fill_value=0
)
```

```python
from math import comb
from sklearn.metrics.pairwise import cosine_similarity
user_similarity=cosine_similarity(user_post_matrix)

def recommend_posts(user_id,num_recomendations=5):
  user_idx=user_post_matrix.index.get_loc(user_id)
  similar_users=(user_similarity[user_idx])
  similar_users_idx=similar_users.argsort()[::-1][1:]

  recommended_posts=set()
  for similar_user_idx in similar_users_idx:
    similar_user_id=user_post_matrix.index[similar_user_idx]
    user_posts=set(combined_data[combined_data["user_id_y"]==similar_user_id]["post_id"])
    recommended_posts.update(user_posts)

    user_posts=set(combined_data[combined_data["user_id_y"]==user_id]["post_id"])
  final_recommendations=list(recommended_posts-user_posts)
  final_recommendations=final_recommendations[:num_recomendations]
  #titles=combined_data[combined_data["post_id"].isin(final_recommendations)]["title"].unique()
  return list(final_recommendations)
print("Recommended Posts:",recommend_posts(user_id=148,num_recomendations=5))
```

```python
import ast
def flatten_dict(d,parent_key="",sep="_"):
  items=[]
  for k,v in d.items():
    new_key= f"{parent_key}{sep}{k}" if parent_key else k
    if isinstance(v,dict):
      items.extend(flatten_dict(v,new_key,sep=sep).items())
    else:
      items.append((new_key,v))
  return dict(items)

def normalize_category(category_column):
  normalized_categories=[]

  for entry in category_column:
    try:
      if isinstance(entry,str):
        entry=ast.literal_eval(entry)
      flattened=flatten_dict(entry)

      category_str=" ".join([f"{k}:{v}" for k,v in flattened.items()])
      normalized_categories.append(category_str)
    except(ValueError,SyntaxError,AttributeError) as e:
      print(f"error processing entry:{entry}-{e}")
      normalized_categories.append("")
  return normalized_categories

df_allposts["category_norm"]=normalize_category(df_allposts["category"])
df_allposts[["category","category_norm","post_id"]]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words="english")
post_tfidf=tfidf.fit_transform(list(df_allposts["category_norm"]))

post_simiarity=cosine_similarity(post_tfidf)

def recommend_similar_posts(post_id,num_recommendations=5):
  post_idx=df_allposts[df_allposts["post_id"]==post_id].index[0]
  similar_posts=post_simiarity[post_idx].argsort()[::-1]
  similar_posts=similar_posts[1:num_recommendations]
  return df_allposts.iloc[similar_posts]["post_id"].tolist()
print("Similar posts:",recommend_similar_posts(post_id=1338,num_recommendations=5)
)
```
